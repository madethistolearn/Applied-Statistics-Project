To the find the winning percentage in team sports, accounting for ties as well. We will use the following formula: 

Winning Percentage= (Number of Wins + 0.5× Number of Ties/Total Number of Games Played)×100%

This formula is especially useful in sports where ties (draws) can occur and accounts for the fact that a tie represents half of a win in terms of calculating the winning percentage

Furthermore, this formula will enable us to study teams with lowest performance because relatively newer teams have played less games so number of games won will not be a fair parameter

```python
x=0
dataset['Winning%']=0
for x in range(61):
    dataset['Winning%'].iloc[x]= (dataset['WonGames'].iloc[x]+ 0.5* dataset['DrawnGames'].iloc[x])/ dataset['PlayedGames'].iloc[x]
dataset.replace(np.nan, 0)#to replace any nan values that were not able to be calucalute with zero
```

### Write the null hypothesis and alternative hypothesis.

-- Null hypothesis (Ho) : There is no distinction in the average funding values between the two categories, closed and operating.

-- Alternate hypothesis (Ha) :  There is significant distinction in the average funding values between the two categories, closed and operating.


```python
import scipy.stats as stats
# Defining the two samples
sample1 = dataset_copy.loc[dataset_copy['OperatingState'] == 'Operating', 'Funds_in_million']
sample2 = dataset_copy.loc[dataset_copy['OperatingState'] == 'Closed', 'Funds_in_million']

# Performing the z-test for two independent samples
z_statistic, p_value = stats.ttest_ind(sample1, sample2)

# Defining the significance level (alpha)
alpha = 0.05

# Comparing p-value with alpha to make a decision
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the means.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the means.")
```

```python
import scipy.stats as stats
contestants_operating = 332  # Number of operating companies among contestants
winners_operating = 19     # Number of operating companies among winners
contestants = 488  # Total number of contestants
winners = 26             # Total number of winners
alpha = 0.05                # Significance level

# Perform proportions Z-test
z_statistic, p_value = stats.proportions_ztest([contestants_operating, winners_operating], [contestants, winners])

if p_value <= alpha:
    print(f"Since the p-value, {round(p_value, 3)} < {alpha} (alpha), the difference is significant and we reject the Null hypothesis")
else:
    print(f"Since the p-value, {round(p_value, 3)} > {alpha} (alpha), the difference is not significant and we fail to reject the Null hypothesis")
```
